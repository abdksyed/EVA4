```
# Import Libraries
import torch
import torchvision.transforms as transforms
import torchvision.datasets as datasets
import numpy as np
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```
transformations = transforms.Compose([transforms.ToTensor()])
```

```
dataset_bg = datasets.ImageFolder("Data/Dataset/bg_", transform = transformations)
```

```
dataset_depth = datasets.ImageFolder("Data/Dataset/depth_", transform = transformations)
```

```
dataset_fgbg = datasets.ImageFolder("Data/Dataset/fg_bg_", transform = transformations)
```

```
dataset_mask = datasets.ImageFolder("Data/Dataset/mask_", transform = transformations)
```

```
from tqdm import tqdm
def stats(dataset):
    dataloader = torch.utils.data.DataLoader(
        dataset, batch_size=1, shuffle=False, num_workers=1)
    mean = torch.zeros(3)
    std = torch.zeros(3)
    print('Computing...')
    for images in tqdm(dataloader):
        for i in range(3):
            mean[i] += images[0][:, i, :, :].mean()
            std[i] += images[0][:, i, :, :].std()
    mean.div_(len(dataset))
    std.div_(len(dataset))
    return mean, std
```

```
mean, std = stats(dataset_bg)
mean_bg = mean.tolist()
std_bg = std.tolist()
print('\n-------------Statisitics-------------')
print(f'Mean of Background image is {mean_bg} and \nStandard Deviation is {std_bg}')
```

```
mean, std = stats(dataset_fgbg)
mean_fgbg = mean.tolist()
std_fgbg = std.tolist()
print('\n-------------Statisitics-------------')
print(f'Mean of FG_BG Overlay image is {mean_fgbg} and \nStandard Deviation is {std_fgbg}')
```

```
mean, std = stats(dataset_depth)
mean_depth = mean.tolist()
std_depth = std.tolist()
print('\n-------------Statisitics-------------')
print(f'Mean of Depth image is {mean_depth} and \nStandard Deviation is {std_depth}')
```

```
mean, std = stats(dataset_mask)
mean_mask = mean.tolist()
std_mask = std.tolist()
print('\n-------------Statisitics-------------')
print(f'Mean of Mask image is {mean_mask} and \nStandard Deviation is {std_mask}')
```

```
print('****************************** FINAL STATISTICS Of 96x96 Image Dataset ******************************')
print(f'**Background**        : mean:{mean_bg}      std:{std_bg}')
print(f'**FG_BG Overlay**     : mean:{mean_fgbg}   std:{std_fgbg}')
print(f'**Mask of FG**        : mean:{mean_mask}     std:{std_mask}')
print(f'**Depth of FG_BG**    : mean:{mean_depth}  std:{std_depth}')
```
