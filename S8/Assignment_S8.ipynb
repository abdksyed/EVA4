```
from packages.imports import * 
#Importing torch, matplotlib. torch.optim
#Importing custom modules - 
            #data (for train and test data and to display)
            #summary (for summary of Networks)
            #test, train (for training and testing methods)
            #utilis (for diplaying graphs, misclassified images, classwise accuracies)
            #model (for the ResNet18 model)
```

```
train_loader, test_loader = data.loader(128) #Loading CIFAR-10 dataset and splitting into train and test dataloader of batch 128.
data.display(64) #Displaying random 64 images from train dataset.
```

```
summary.summ(device, 'ResNet18') #To display netwrok archeiture and parameters used 
```

```
model = Net.ResNet18().to(device) #Creadting an instance named model, for the ResNet18 and converting to CUDA.
optimizer = optim.SGD(model.parameters(), lr = 0.01, momentum= 0.9, weight_decay= 1e-5) #Defining Stochastic Gradient Descent Optimizer

epochs = 50 #Defining number of epochs

for epoch in range(epochs): #Looping over each epoch to train and test model
  print('EPOCH: ', epoch+1)
  train.train(model, device, train_loader, optimizer, epoch)
  test.test(model, device, test_loader)
```
